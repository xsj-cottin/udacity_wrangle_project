### 1.3 Use the Tweepy lAPI to query additional data via the Twitter API (tweet_json.txt). This was done in two steps to verify functionality.

<a id="tweeter"></a>

First, the data for only one tweet id was retieved to figure out how to handle the tweet API status object and then the tweet API user object

A lot of googling took place to figure those out but two that come to mind are:

https://www.geeksforgeeks.org/python-status-object-in-tweepy/

dealing with API errors
https://stackoverflow.com/questions/6817640/catch-any-error-in-python
    
Dealing with rate limit
https://stackoverflow.com/questions/37248958/tweepy-wait-on-rate-limit-not-working
    
Then a for loop was created to run all the ids.
Those articles above also prompted the way the code is written to nullify the need for the JSON data when all that is required here can be simply extracted from the status object. However, to agree with the project's requirements the JSON data was also downloaded.

NOTE:
Since we wants only tweets that have an image, we use the list of tweet ids from df2 NOT that from df1 since df2 has the pictures' URL

```python
import tweepy
import json
    
tweet_ids=df2['tweet_id']

good_id=[]
full_text=[]
retweet_count=[]
favorite_count=[]
time_stamp=[]
tweet_error={'Tweet_id':[],'Error':[]}

consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify =True)

with open('tweet_json.txt','w', newline='\n') as f, open('tweet_jason_user.txt','w', newline='\n') as g:
    for id in tweet_ids:
        try:
            status=api.get_status(id,tweet_mode='extended')
        except tweepy.TweepError as e: 
            tweet_error['Tweet_id'].append(id)
            tweet_error['Error'].append(e.response.text)
            continue
            
        good_id.append(id)   
        full_text.append(status.full_text)
        time_stamp.append(status.created_at)
        retweet_count.append(status.retweet_count)
        favorite_count.append(status.favorite_count)
        json.dump(status._json,f)
        f.write('\n')
        json.dump(status.user._json,g)
        g.write('\n')
        
        
print("DONE")
```
